# Image Classification

Image Classification : For an input image, assigned a label (from a set of known labels)

Semantic Gap : Gap between pixel values in an image and the semantic idea present in the image (e.g. a cat)



#### Challenges
- Viewpoint variation
- Different illumination for same object
- Deformation
- Occlusion (object getting obstructed by other things)
- Background clutter (object and background are difficult to separate)
- Intraclass variation (e.g. need to identify all species of cats)


# Nearest Neighbour Classifier
- Find the image closest to input image in training data.
- Training is O(1), prediction required O(N) time.
- Fast to train, slow to predict (want it to be opposite).
- Sensitive to outliers.



# K-Nearest Neighbours
- Depend on majority vote from 'k' nearest neighbours.
- k : A hyper-parameter as it cannot be inferred from the data. Need to experiment to decide this.


### Distance Metrics :
- L1 Distance : Dependent on choice of co-ordinate system (CS).
- L2 Distance : Slow but independent of CS.

# Choosing Hyperparameters


- **Train Set** : Subset of data used to train the model.
- **Validation Set** : Subset of data used to assess model performance and choose a model and its hyper-parameters.
- **Test Set** : Subset of data used to estimate model performance on unseen, real-world data. This performance is reported.


Test set separation is important so that model chosen is not biased towards the dataset available but also generalizes well to unknown data.

# Limitations of kNN

- Very slow at prediction time (need to search training data space for closest neighbours)
- Curse of dimensionality : Need lot of data (exponential in the number of the dimensions) to densely fill the high-dimensional space so that nearest neighbours are not very far from input data. If the training space is very sparse, the nearest neighbours can be images very different from the input. Also search space increases exponentially.



# Parametric Approach : Linear Classifiers
- Try to condense knowledge from training data into some fixed number of parameters/weights.

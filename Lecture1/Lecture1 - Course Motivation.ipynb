# Brief History of Vision (and computer vision)

### Biological Vision
- The first animals developed vision about 540 million year ago.
- This started an evolutionary arms race and a lot of animals developed vision very quickly.

### Camera
- First cameras were pinhole cameras (camera obscura).

### Studies on mechanism of vision
#### Hubel and Wiesel, 1959
- Studied cat brains using electrodes.
- Found that cells in visual cortex respond to simple structures like oriented edges.
- Visual processing starts using simple structures like oriented edges and build on top of simple structures to understand visual information.

#### The Summer Vision Project, 1966
- Goal was to understand vision in 1 summer.
- Unfulfilled till now !

#### Stages of Visual Representation by David Marr


- Visual understanding comprises of several stages (primal sketch, 2.5D sketch and then 3D understanding)

#### Generalized Cylinder and Pictorial Structure


- Idea : Every object is combosed of simple geometrical primitives. Reduces complex structure to simpler shapes.

### Image DataSets
- Image data is generally very high-dimensional data.
- If we do not provide a lot of training data, overfitting can happen very easily and the model will not generalize to unseen data.
- So scientists started preparing labelled image datasets for training and benchmarking.
- ImageNet (www.image-net.org) is one of the largest datasets in image recognition.



- Error rates have been steadily decreasing, to almost human levels.
- Around 2012, error rates dropped drastically due to use of CNNs.
- However not all problems of image understanding have been solved yet.

# Visual Recognition Problems

- Object detection - Identify several objects in one single image and draw bounding boxes around them.
- Image captioning - Produce a natural language sentence describing the image.
- Image Segmentation 
- Activity recognition - What is the person doing in the video ?


# Convolutional Neural Networks



- Neural networks have been getting deeper every year.
- CNNs are not a new idea though. Were first introduced in 1998 (LeCun et al)
- Why did CNN suddenly become popular ?
    - Increase in computational power (specially GPUs)
    - CNNs are very data hungry. Due to internet, collecting large amounts of labelled data became possible.

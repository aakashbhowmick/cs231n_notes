# Motivation
- Amount of images have exploded. Mostly due to increase in number of sensors (specially in smartphones)
- Majority of bits on internet are visual data.
- Very useful to be able to make sense of this visual data.
- But it is very difficult for a machine to understand visual data.
- For streaming websites like YouTube where > 5 hours of video is uploaded every 1s, it is impossible for humans to keep up with this scale.
